```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Data
measurements = pd.read_csv('/content/realistic_ocean_climate_dataset.csv')
measurements.head()
```

```python
# Visualizing the Data

# Median Values
# group the measurements by Location
# and find the median values for pH Level and SST(°C)
grouped_measurements = measurements.groupby('Location')[['pH Level', 'SST (°C)']].median()
print(grouped_measurements)
```

```python
# Replace 'nan' string with actual NaN and fill any NaN with 0
measurements['Bleaching Severity'] = measurements['Bleaching Severity'].replace('nan', pd.NA).fillna(0)

# Map the severity levels to integers
severity_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
measurements['Bleaching Severity'] = measurements['Bleaching Severity'].replace(severity_mapping)

# Convert the column to integer type
measurements['Bleaching Severity'] = measurements['Bleaching Severity'].astype(int)

print(measurements['Bleaching Severity'].dtype)
print(measurements['Bleaching Severity'].value_counts())
```

```python
# Visualizing the Data

# Median Values
# group the measurements by Location
# and find the median values for pH Level and SST(°C)
grouped_measurements = measurements.groupby('Location')[['pH Level', 'SST (°C)']].median()
print(grouped_measurements)
```

```python
# plot the column chart. set the figsize to (8,4)
ax_bar = grouped_measurements.plot(kind='bar')

# customize the plot
# set the x-axis label
plt.xlabel('Location', fontsize = 14)

# set the y-axis label
plt.ylabel('Values', fontsize = 14)

# set the title
plt.title('Median pH Level and SST', fontsize = 16)

# remove top spine
ax_bar.spines['top'].set_visible(False)

# remove right spine
ax_bar.spines['right'].set_visible(False)

# show the plot
plt.show()



```

```python
# Correlation Between Measurements
# Correlation between pH Level and Species Observed in Hawaiian Islands
Hawaiian_measurements = measurements[measurements['Location']=='Hawaiian Islands']

plt.figure()

# create the scatterplot
pH_Species_scatter = plt.scatter(x=Hawaiian_measurements['pH Level'], y=Hawaiian_measurements['Species Observed'], s=10, marker='s', color= 'darkorange', alpha=0.6)

# set the title
plt.title('Correlation between pH Level and Species Observed in Hawaiian Islands', fontsize = 16)

# set the x-axis label
plt.xlabel('pH Level', fontsize = 14)

# set the y-axis label
plt.ylabel('Species Observed', fontsize = 14)

# show the plot
plt.show()
```

```python
print("The correlation between pH Level and Species Observed for the Hawaiian Island is:", round(Hawaiian_measurements["pH Level"].corr(Hawaiian_measurements["Species Observed"]),4))
```

```python
# Hawaiian Histogram
# Species observed in Hawaii

plt.figure()

# create the histogram

Hawaiian_hist_ax = sns.histplot(Hawaiian_measurements,x='Species Observed',binwidth=5,kde=True,color='royalblue',alpha=0.2)

# set the xlabel
plt.xlabel('Species Observed', fontsize = 14)

# set the ylabel
plt.ylabel('Frequency', fontsize = 14)

# set the title
plt.title('Species Observed in Hawaii', fontsize = 16)

# add grid on the y axis
plt.grid(axis='y', linestyle=':', alpha=0.7)

plt.show()
```

```python
# Multiple Histograms

plt.figure()

# select the subplot for the SST (°C) histogram
hist_ax1 = plt.subplot(2, 1, 1)

# create the histogram for SST (°C)
sns.histplot(measurements,
    x='SST (°C)',
    hue='Bleaching Severity')

# set the xlabel
hist_ax1.set_xlabel('SST (°C))', fontsize=14)

# set the title
hist_ax1.set_title('Distribution of the SST (°C) with Bleaching Severity', fontsize=16)


# Repeat for Species Observed histogram
# select the subplot for the Species Observed histogram
hist_ax2 = plt.subplot(2, 1, 2)

# create the histogram for Species Observed
sns.histplot(measurements,
    x='Species Observed',
    hue='Bleaching Severity')

# set the xlabel
hist_ax2.set_xlabel('Species Observed', fontsize=14)

# set the title
hist_ax2.set_title('Distribution of Species Observed with Bleaching Severity', fontsize=16)

### END CODE HERE ###

plt.tight_layout()
plt.show()
```

```python
# Box plots

boxplot_fig = plt.figure()

# create the boxplot
boxplot_ax= sns.boxplot(data=measurements, x='Species Observed', hue='Bleaching Severity')

# set the xlabel
boxplot_ax.set_xlabel('Species Observed', fontsize=14)

# set the ylabel
boxplot_ax.set_ylabel('Bleaching Severity', fontsize=14)

# set the title
boxplot_ax.set_title('Boxplot of Species Observed by Bleaching Severity', fontsize=16)

plt.show()
```
